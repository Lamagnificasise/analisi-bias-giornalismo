```python
# 1. Questa è la nostra frase "sporca"
testo_grezzo = "Il MAROCCHINO, fermato ieri sera... è stato rilasciato."

# 2. Trasformiamo tutto in minuscolo
testo_minuscolo = testo_grezzo.lower()

# 3. Togliamo la punteggiatura (usiamo una funzione semplice)
import string
testo_senza_punti = testo_minuscolo.translate(str.maketrans('', '', string.punctuation))

print("Testo pulito:", testo_senza_punti)
```

```python
import nltk
from nltk.corpus import stopwords

# Scarichiamo la lista delle parole inutili (il, lo, la, di, su...)
nltk.download('stopwords')
stop_words_italiane = set(stopwords.words('italian'))

# La nostra frase di prima (già pulita da punteggiatura e minuscola)
frase_pulita = "il marocchino ieri è stato rilasciato dalla polizia"

# Dividiamo la frase in singole parole
parole = frase_pulita.split()

# Teniamo solo le parole che NON sono nella lista delle stopwords
risultato = [p for p in parole if p not in stop_words_italiane]

print("Prima:", frase_pulita)
print("Dopo la pulizia profonda:", risultato)
```

```python
def trova_vicini(lista_parole, target, raggio=2):
    vicini = []
    for i, parola in enumerate(lista_parole):
        if parola == target:
            # Prende le parole prima e dopo nel raggio scelto
            inizio = max(0, i - raggio)
            fine = min(len(lista_parole), i + raggio + 1)
            # Aggiunge i vicini alla nostra lista (escludendo il target stesso)
            contestuali = lista_parole[inizio:fine]
            contestuali.remove(target)
            vicini.extend(contestuali)
    return vicini

# Testiamolo!
testo_test = ["polizia", "arresta", "marocchino", "per", "furto", "mentre", "egiziano", "aiuta", "anziana"]
risultato_vicini = trova_vicini(testo_test, "marocchino")

print(f"Vicini di 'marocchino': {risultato_vicini}")
```
